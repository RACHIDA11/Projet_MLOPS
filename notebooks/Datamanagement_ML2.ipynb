####3.3.1/Convertir les valeurs de MW en MWh

Dans le cadre de l'analyse des données de consommation et de production énergétique, il est essentiel de convertir les valeurs mesurées en mégawatts (MW) en mégawattheures (MWh) pour obtenir une représentation précise de l'énergie produite ou consommée sur une période donnée. Les valeurs en MW représentent la puissance à un instant donné, tandis que les MWh mesurent l'énergie totale sur une durée spécifique. Étant donné que les données sont collectées toutes les 30 minutes pour le df_4 et toutes les 15 min pour le df_3, chaque valeur en MW doit être multipliée par 0,5 ou 0,24  pour obtenir les valeurs en MWh

```python
unique_heures_df_3 = df_3['Heure'].unique()
print("Valeurs uniques des heures dans df_3 :", unique_heures_df_3)
```

```python
unique_heures_df_4 = df_4['Heure'].unique()
print("Valeurs uniques des heures dans df_4 :", unique_heures_df_4)
```

```python



# Convertir les colonnes en types numériques
cols_to_convert = ['Consommation (MW)', 'Thermique (MW)', 'Nucléaire (MW)',
                   'Eolien (MW)', 'Solaire (MW)', 'Hydraulique (MW)', 'Bioénergies (MW)']

for col in cols_to_convert:
    df_3[col] = pd.to_numeric(df_3[col], errors='coerce')
    df_4[col] = pd.to_numeric(df_4[col], errors='coerce')

# Fonction pour convertir MW en MWh en fonction de la durée
def convertir_en_mwh(df, colonne_mw, nouvelle_colonne, duree_en_heures):
    df[nouvelle_colonne] = df[colonne_mw] * duree_en_heures

# Ajouter des colonnes pour df_3 (15 minutes = 0.25 heures)
convertir_en_mwh(df_3, 'Consommation (MW)', 'Consommation (MWh)', 0.25)
convertir_en_mwh(df_3, 'Thermique (MW)', 'Thermique (MWh)', 0.25)
convertir_en_mwh(df_3, 'Nucléaire (MW)', 'Nucléaire (MWh)', 0.25)
convertir_en_mwh(df_3, 'Eolien (MW)', 'Eolien (MWh)', 0.25)
convertir_en_mwh(df_3, 'Solaire (MW)', 'Solaire (MWh)', 0.25)
convertir_en_mwh(df_3, 'Hydraulique (MW)', 'Hydraulique (MWh)', 0.25)
convertir_en_mwh(df_3, 'Bioénergies (MW)', 'Bioénergies (MWh)', 0.25)

# Ajouter des colonnes pour df_4 (30 minutes = 0.5 heures)
convertir_en_mwh(df_4, 'Consommation (MW)', 'Consommation (MWh)', 0.5)
convertir_en_mwh(df_4, 'Thermique (MW)', 'Thermique (MWh)', 0.5)
convertir_en_mwh(df_4, 'Nucléaire (MW)', 'Nucléaire (MWh)', 0.5)
convertir_en_mwh(df_4, 'Eolien (MW)', 'Eolien (MWh)', 0.5)
convertir_en_mwh(df_4, 'Solaire (MW)', 'Solaire (MWh)', 0.5)
convertir_en_mwh(df_4, 'Hydraulique (MW)', 'Hydraulique (MWh)', 0.5)
convertir_en_mwh(df_4, 'Bioénergies (MW)', 'Bioénergies (MWh)', 0.5)

# Afficher les premières lignes des DataFrames pour vérifier les nouvelles colonnes
print(df_3.head())
print(df_4.head())
```

###4.1/ Variables météo :Rayonnement solaire et vent et temp: 	Imputer par la médiane (un jour avant un jour après / 2 jours avant et 2 jours après)

```python
# Liste des colonnes météo à traiter (température, rayonnement solaire, vitesse du vent)
meteo_columns = [
    "TMin (°C)", "TMax (°C)", "TMoy (°C)",
    "Rayonnement solaire global (W/m2)", "Vitesse du vent à 100m (m/s)"
]

# Appliquer l'imputation dans l'ordre : 1 jour avant, 1 jour après
for column in meteo_columns:
    df_full[column] = df_full.groupby("Code INSEE région")[column].transform(
        lambda x: x.fillna(
            # D'abord imputer avec 1 jour avant et 1 jour après
            pd.concat([x.shift(1), x.shift(-1)]).median()
        )
    )

    # Ensuite, une fois que les NaN sont remplis avec les 1 jour avant et après, on utilise 2 jours avant et après
    df_full[column] = df_full.groupby("Code INSEE région")[column].transform(
        lambda x: x.fillna(
            # Puis imputer avec 2 jours avant et 2 jours après (si encore des NaN après la première étape)
            pd.concat([x.shift(2), x.shift(-2)]).median()
        )
    )

# Vérifier le nombre de valeurs manquantes dans chaque colonne après l'imputation
for column in meteo_columns:
    missing_values = df_full[column].isna().sum()
    print(f"Nombre de valeurs manquantes dans la colonne '{column}': {missing_values}")
```

Nb.Le code d'abord remplit les valeurs manquantes avec les jours les plus proches (1 jour avant et 1 jour après).
Si certaines valeurs restent manquantes après cette première étape, il passe à l'utilisation des données des 2 jours avant et après.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import holidays

# Supposons que df_full_filtered est déjà défini et que la colonne 'Date' est au format datetime

# Assurez-vous que la colonne 'Date' est au format datetime
df_full_filtered['Date'] = pd.to_datetime(df_full_filtered['Date'])

# Ajouter des variables temporelles
df_full_filtered['Jour'] = df_full_filtered['Date'].dt.day
df_full_filtered['Mois'] = df_full_filtered['Date'].dt.month
df_full_filtered['Année'] = df_full_filtered['Date'].dt.year
df_full_filtered['Semaine_de_l_année'] = df_full_filtered['Date'].dt.isocalendar().week
df_full_filtered['Jour_de_la_semaine'] = df_full_filtered['Date'].dt.dayofweek  # 0 = Lundi, 6 = Dimanche

# Ajouter une colonne pour les jours fériés
fr_holidays = holidays.France(years=[2016, 2023])  # Créer un objet pour les jours fériés en France
df_full_filtered['Jour_férié'] = df_full_filtered['Date'].dt.date.isin(fr_holidays).astype(int)

# Ajout de transformations sinus et cosinus pour la semaine de l'année
df_full_filtered['Semaine_sin'] = np.sin(2 * np.pi * df_full_filtered['Semaine_de_l_année'] / 52)
df_full_filtered['Semaine_cos'] = np.cos(2 * np.pi * df_full_filtered['Semaine_de_l_année'] / 52)

# Représentation trigonométrique pour "Jour" (1 à 31) : sin et cos
df_full_filtered['Jour_sin'] = np.sin(2 * np.pi * df_full_filtered['Jour'] / 31)
df_full_filtered['Jour_cos'] = np.cos(2 * np.pi * df_full_filtered['Jour'] / 31)

# Représentation trigonométrique pour "Mois" (1 à 12) : sin et cos
df_full_filtered['Mois_sin'] = np.sin(2 * np.pi * df_full_filtered['Mois'] / 12)
df_full_filtered['Mois_cos'] = np.cos(2 * np.pi * df_full_filtered['Mois'] / 12)

# Vérifiez les premières lignes de df_full_filtered pour vous assurer que tout est correct
print(df_full_filtered.head())
```
